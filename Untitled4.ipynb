```python
import pandas as pd
import pandas.io.sql as sqlio
import psycopg2 as ps

conn2 = ps.connect(
    dbname = "Parch",
    user = "postgres",
    password = "abc123",
    host = "localhost",
    port = "5432" )

sql = """
SELECT
    DATE(occurred_at) AS event_date,
    channel AS event_channel,
    COUNT(id) AS event_count
FROM
    web_events
GROUP BY
    event_date,
    event_channel
ORDER BY
    event_date,
    event_channel;


"""
df = sqlio.read_sql_query(sql,conn2)
df.head(5)
```
