```python
# Q - 5

def NumPy_Format(arr):
    
    column_widths = [max(len(str(arr[i][j])) for i in range(len(arr))) for j in range(len(arr[0]))]

    Result_Array = ""

    for row in arr:
        for i, element in enumerate(row):
            
            spaces = column_widths[i] - len(str(element))
            
            Result_Array += " " * spaces + str(element)
            if i < len(row) - 1:
                Result_Array += "  "
        Result_Array += "\n"  

    return Result_Array


array = [ [1, 12, 1523], [425, 66, 7849],[12, 1237, 56]]

Result_Array = NumPy_Format(array)
print(Result_Array)
```

```python
# Q - 6

import pandas as pd


def Pandas_Format(df):


    maximum_col_widths = {}
    for col in df.columns:
        maximum_col_widths[col] = max([len(str(val)) for val in df[col].tolist()])

    Total_rows = len(df)
    Total_columns = len(df.columns)

    output = ""
    for row_index in range(Total_rows + 1):
        for column_index in range(Total_columns + 1):
            if row_index == 0 and column_index == 0:
                val = " "
            elif row_index == 0:
                val = df.columns[column_index - 1]
            elif column_index == 0:
                val = df.index[row_index - 1]
            else:
                val = df.iloc[row_index - 1, column_index - 1]
            width = maximum_col_widths[df.columns[column_index - 1]] if column_index > 0 else 0
            output += f"{str(val):>{width}}  "
        output += "\n"

    return output


df = pd.DataFrame([[41, 32, 53], [43, 53, 56], [67, 78, 89]], columns=["A", "B", "C"])
print(Pandas_Format(df))
```

```python
# Q - 7


import numpy as np


board_Array = np.array([['-', '-', '-'], ['-', '-', '-'], ['-', '-', '-']])


def create_board():
    print(board_Array[0])
    print(board_Array[1])
    print(board_Array[2])


def check_win(player):
    
    for i in range(3):
        if np.all(board_Array[i] == player):
            return True
   
    for i in range(3):
        if np.all(board_Array[:,i] == player):
            return True
  
    if np.all(np.diagonal(board_Array) == player) or np.all(np.diagonal(np.fliplr(board_Array)) == player):
        return True
    return False


def check_draw():
    return '-' not in board_Array

print("====================================Tic-Tac-Toe Game===============================================")
create_board()
player = 'X'
while True:
    
    while True:
        row = input(f"Player {player}, enter row  ")
        col = input(f"Player {player}, enter col  ")
        row, col = int(row), int(col)
        
        if board_Array[row][col] == '-':
            break
        else:
            print("The Square is Marked by other Player . Try Another Square")
    
    board_Array[row][col] = player
    create_board()
   
    if check_win(player):
        print(f"Player {player} wins the game")
        break
    if check_draw():
        print("The game is a drawn")
        break
    
    if player == 'X':
        player = 'O'
    else:
        player = 'X'
```

```python
#  Q - 8
import numpy as np
import random

board = np.array([['-', '-', '-'], ['-', '-', '-'], ['-', '-', '-']])

def Create_board():
    print(board[0])
    print(board[1])
    print(board[2])

def Winner(player):
    for i in range(3):
        if np.all(board[i] == player):
            return True

    for i in range(3):
        if np.all(board[:, i] == player):
            return True

    if np.all(np.diag(board) == player) or np.all(np.diag(np.fliplr(board)) == player):
        return True
    return False

def check_draw():
    return '-' not in board

def player_turn(player):
    while True:
        row = input(f"Player {player}, enter row  ")
        col = input(f"Player {player}, enter col  ")
        row, col = int(row), int(col)
        
        if board[row][col] == '-':
            return row, col
        else:
            print("The Square is Marked by other Player . Try Another Square")
    

def computer_turn(player):
    print(f"System Turn")
    Moves_Possible = [(r, c) for r in range(3) for c in range(3) if board[r][c] == '-']
    return random.choice(Moves_Possible)

def main():
    print("====================================Tic-Tac-Toe Game===============================================")
    Create_board()

    player = 'X'
    computer = 'O'

    while True:
        if player == 'X':
            row, col = player_turn(player)
        else:
            row, col = computer_turn(player)

        board[row][col] = player
        Create_board()

        if Winner(player):
            if player == 'X':
                print("Player X wins!")
            else:
                print("Computer  wins")
            break

        if check_draw():
            print("The game is a drawn.")
            break

        if player == 'X':
            player = 'O'
        else:
            player = 'X'

if __name__ == "__main__":
    main()

```

```python
# Q-12

import numpy as np

def Calculate_Median(array):
    array = array.flatten()
    array = np.sort(array)
  
    if len(array) % 2 == 0:
        return (array[int(len(array)/2)] + array[int(len(array)/2) - 1]) / 2
    else:
        return array[int(len(array)/2)]

def Calculate_Mode(array):
    
    
    values, counts = np.unique(array, return_counts=True)
    index = np.argmax(counts)
    return values[index]


arr1 = np.array([11, 22, 35, 46, 5])

arr2 = np.array([[1, 24, 3], [47, 53, 6]])

arr3 = np.array([1, 22, 3, 43, 4, 44, 4, 5, 53, 53, 45])


print("\nArray 1 Values \n")
print(arr1)
print("\nMedian of Array 1 \n")
print(Calculate_Median(arr1))
print("\nCalculate_Mode of Array 1 \n")
print(Calculate_Mode(arr1))


print("\nArray 2 Values \n")
print(arr2)
print("\nMedian of Array 2 \n")
print(Calculate_Median(arr2))
print("\nCalculate_Mode of Array 2 \n")
print(Calculate_Mode(arr2))

print("\nArray 3 Values \n")
print(arr3)
print("\nMedian of Array 3 \n")
print(Calculate_Median(arr3))
print("\nCalculate_Mode of Array 3 \n")
print(Calculate_Mode(arr3))

```

```python
# Q 13

import numpy as np

def enhanced_median(arr, axis=None):
    
    arr = np.array(arr)
    
    if axis is None:
        return np.median(arr)
    elif axis == 0:
        return np.median(arr, axis=0)
    elif axis == 1:
        return np.median(arr, axis=1)
    else:
        raise ValueError("Axis 0 = column and axis 1 = row")

def enhanced_mode(arr, axis=None):
    
    arr = np.array(arr)
    
    if axis is None:
        return np.apply_along_axis(lambda x: np.argmax(np.bincount(x)), axis=0, arr=arr)
    elif axis == 0:
        return np.apply_along_axis(lambda x: np.argmax(np.bincount(x)), axis=0, arr=arr)
    elif axis == 1:
        return np.apply_along_axis(lambda x: np.argmax(np.bincount(x)), axis=1, arr=arr)
    else:
        raise ValueError("Axis 0 = column and axis 1 = row")

array1 = np.array([[61, 24, 33, 44],
                [53, 2, 32, 45],
                [15, 25, 33, 42]])

array2 = np.array([[13, 42, 33, 42],
                [32, 22, 13, 24],
                [15, 25, 43, 34]])

array3 = np.array([[17, 32, 33, 42],
                [32, 24, 11, 24],
                [75, 27, 44, 54]])

print("Array 1\n")

column_medians = enhanced_median(array1, axis=0)
print("Median in Column", column_medians)

row_medians = enhanced_median(array1, axis=1)
print("Median in Row", row_medians)

column_modes = enhanced_mode(array1, axis=0)
print("Mode in Column", column_modes)

row_modes = enhanced_mode(array1, axis=1)
print("Mode in Row", row_modes)

print("\nArray 2\n")

column_medians = enhanced_median(array2, axis=0)
print("Median in Column", column_medians)

row_medians = enhanced_median(array2, axis=1)
print("Median in Row", row_medians)

column_modes = enhanced_mode(array2, axis=0)
print("Mode in Column", column_modes)

row_modes = enhanced_mode(array2, axis=1)
print("Mode in Row", row_modes)

print("\nArray 3\n")
column_medians = enhanced_median(array3, axis=0)
print("Median in Column", column_medians)

row_medians = enhanced_median(array3, axis=1)
print("Median in Row", row_medians)

column_modes = enhanced_mode(array3, axis=0)
print("Mode in Column", column_modes)

row_modes = enhanced_mode(array3, axis=1)
print("Mode in Row", row_modes)
```

```python
#Q-14 
#B

import numpy as np

board = np.zeros((8, 8), dtype=int)

horizontal = []
vertical = []

horizontal = [2, 1, -1, -2, -2, -1, 1, 2]
vertical = [-1, -2, -2, -1, 1, 2, 2, 1]

current_row, current_column = 0, 0
move_count = 1

while move_count <= 64:
    board[current_row][current_column] = move_count
    move_count += 1

    best_move = -1
    min_moves = 9

    for move_number in range(8):
        new_row = current_row + vertical[move_number]
        new_col = current_column + horizontal[move_number]

        if 0 <= new_row < 8 and 0 <= new_col < 8 and board[new_row][new_col] == 0:
            valid_moves = 0
            for next_move in range(8):
                next_row = new_row + vertical[next_move]
                next_col = new_col + horizontal[next_move]
                if 0 <= next_row < 8 and 0 <= next_col < 8 and board[next_row][next_col] == 0:
                    valid_moves += 1

            if valid_moves < min_moves:
                min_moves = valid_moves
                best_move = move_number

    current_row += vertical[best_move]
    current_column += horizontal[best_move]


total_moves = np.max(board)

for row in board:
    for count in row:
        print(f'{count:2}', end=' ')
    print()


print(f"The knight game made {total_moves} moves to travel all the 64  squares.")
```
