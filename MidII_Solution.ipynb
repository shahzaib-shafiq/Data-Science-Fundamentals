```python
import pandas as pd
```

```python
unames=['user_id','gender','age', 'occupation','zip']
users = pd.read_csv(r'./datasets/ml-100k/users.dat', sep="::", names = unames, engine="python")
users
```

```python
rnames = ["user_id", "movie_id", "rating", "timestamp"]
ratings = pd.read_csv(r'./datasets/ml-100k/ratings.dat', sep='::', names=rnames, engine="python")
ratings
```

```python
mnames = ["movie_id", "title", "genre"]
movies = pd.read_csv(r'./datasets/ml-100k/movies.dat', sep='::', names=mnames, engine="python")
movies
```

```python
df = pd.merge(ratings, users, on="user_id")
df
data = pd.merge(df, movies, on="movie_id")
```

```python
data
```

```python
data["gender"].value_counts()
```

```python
mean_ratings = data.pivot_table(index="title", columns="gender", values="rating", aggfunc="mean")
mean_ratings
```

```python
top_female_ratings = mean_ratings["F"].sort_values(ascending=False)
top_female_ratings.head(5)
```

```python
df = data[["title", "rating"]]
title_profile = df.groupby("title").count()
title_profile


filtered_title = title_profile[title_profile["rating"]>250]
filtered_title.sort_values("rating",ascending=True)
```

```python
#the sliced data is in the df
df = data[["title", "genre"]]
df
genre_profile = df.groupby("genre").count() 
genre_profile
```
