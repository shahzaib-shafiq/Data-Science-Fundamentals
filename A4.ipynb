```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#%matplotlib inline
```

```python
data = pd.read_csv('fatalities.csv')
```

```python


injury_types = data['type_of_injury']


injury_count = injury_types.value_counts()
#creating plot

plt.figure(figsize=(10, 6))
injury_count.plot(kind='bar', color='skyblue')
plt.title('Frequency of Injury Types')
plt.xlabel('Injury Types')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
injury_count = injury_types.value_counts()

# Plotting a pie chart to show the distribution of injury types
plt.figure(figsize=(8, 8))
plt.pie(injury_count, labels=injury_count.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Injury Types')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```

```python
#Q-5



ammunition_frequency = data['ammunition'].value_counts()
means_of_killing_frequency = data['killed_by'].value_counts()


plt.figure(figsize=(10, 6))
ammunition_frequency.head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Most Frequent Ammunition Used')
plt.xlabel('Ammunition')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# Plotting the most frequent means of killing
plt.figure(figsize=(10, 6))
means_of_killing_frequency.head(10).plot(kind='bar', color='salmon')
plt.title('Top 10 Most Frequent Means of Killing')
plt.xlabel('Means of Killing')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python

ammunition_counts = data['ammunition'].value_counts()

# Count the occurrences of means by which individuals were killed
killed_by_counts = data['killed_by'].value_counts()

# Plotting the most frequently used ammunition
plt.figure(figsize=(10, 6))
ammunition_counts.head(10).plot(kind='bar', color='skyblue')
plt.title('Most Frequently Used Ammunition')
plt.xlabel('Ammunition')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# Plotting the most frequently used means of killing
plt.figure(figsize=(10, 6))
killed_by_counts.head(10).plot(kind='bar', color='lightgreen')
plt.title('Most Frequently Used Means of Killing')
plt.xlabel('Means of Killing')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
