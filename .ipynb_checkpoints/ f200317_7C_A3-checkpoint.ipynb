```python
# Q - 5

def NumPy_Format(arr):
    
    column_widths = [max(len(str(arr[i][j])) for i in range(len(arr))) for j in range(len(arr[0]))]

    formatted_output = ""

    for row in arr:
        for i, element in enumerate(row):
            
            spaces = column_widths[i] - len(str(element))
            
            formatted_output += " " * spaces + str(element)
            if i < len(row) - 1:
                formatted_output += "  "
        formatted_output += "\n"  

    return formatted_output


array = [ [1, 12, 1523], [425, 66, 7849],[12, 1237, 56]]

formatted_output = NumPy_Format(array)
print(formatted_output)
```

```python
# Q - 6

import pandas as pd


def format_dataframe(df):


    max_col_widths = {}
    for col in df.columns:
        max_col_widths[col] = max([len(str(val)) for val in df[col].tolist()])

    num_rows = len(df)
    num_cols = len(df.columns)

    output = ""
    for row_index in range(num_rows + 1):
        for column_index in range(num_cols + 1):
            if row_index == 0 and column_index == 0:
                val = " "
            elif row_index == 0:
                val = df.columns[column_index - 1]
            elif column_index == 0:
                val = df.index[row_index - 1]
            else:
                val = df.iloc[row_index - 1, column_index - 1]
            width = max_col_widths[df.columns[column_index - 1]] if column_index > 0 else 0
            output += f"{str(val):>{width}}  "
        output += "\n"

    return output


df = pd.DataFrame([[41, 32, 53], [43, 53, 56], [67, 78, 89]], columns=["A", "B", "C"])
print(format_dataframe(df))
```

```python
#  Q - 8
import numpy as np
import random

board = np.array([['-', '-', '-'], ['-', '-', '-'], ['-', '-', '-']])

def print_board():
    print(board[0])
    print(board[1])
    print(board[2])

def check_win(player):
    for i in range(3):
        if np.all(board[i] == player):
            return True

    for i in range(3):
        if np.all(board[:, i] == player):
            return True

    if np.all(np.diag(board) == player) or np.all(np.diag(np.fliplr(board)) == player):
        return True
    return False

def check_draw():
    return '-' not in board

def player_turn(player):
    while True:
        move = input(f"Player {player}, enter your move (row, column): ")
        row, col = move.split(',')
        row, col = int(row), int(col)
        if board[row][col] == '-':
            return row, col
        else:
            print("That square is already taken. Try again.")

def computer_turn(player):
    print(f"computer is making move")
    available_moves = [(r, c) for r in range(3) for c in range(3) if board[r][c] == '-']
    return random.choice(available_moves)

def main():
    print("====================================Tic-Tac-Toe Game===============================================")
    print_board()

    player_choice = input("Do you want to go first (Y/N)? ").strip().lower()
    if player_choice == 'y':
        player = 'X'
        computer = 'O'
    else:
        player = 'O'
        computer = 'X'

    while True:
        if player == 'X':
            row, col = player_turn(player)
        else:
            row, col = computer_turn(player)

        board[row][col] = player
        print_board()

        if check_win(player):
            if player == 'X':
                print("Player X wins!")
            else:
                print("Computer  wins")
            break

        if check_draw():
            print("The game is a draw.")
            break

        if player == 'X':
            player = 'O'
        else:
            player = 'X'

if __name__ == "__main__":
    main()

```

```python
# Q - 9

import random

def print_board(board):
    for i in range(4):
        for j in range(4):
            print(" ".join(board[i][j]))
        print("-" * 17)

def check_win(board, player):
    # Check for wins in each 2D plane
    for i in range(4):
        for j in range(4):
            if all(board[i][j][k] == player for k in range(4)):
                return True

    for i in range(4):
        for k in range(4):
            if all(board[i][j][k] == player for j in range(4)):
                return True

    for j in range(4):
        for k in range(4):
            if all(board[i][j][k] == player for i in range(4)):
                return True

    # Check for wins in 2D diagonals
    for i in range(4):
        if all(board[i][j][j] == player for j in range(4)):
            return True
        if all(board[i][j][3 - j] == player for j in range(4)):
            return True

    for j in range(4):
        if all(board[i][j][i] == player for i in range(4)):
            return True
        if all(board[i][j][3 - i] == player for i in range(4)):
            return True

    for k in range(4):
        if all(board[i][i][k] == player for i in range(4)):
            return True
        if all(board[i][3 - i][k] == player for i in range(4)):
            return True

    # Check for wins in 3D diagonals
    if all(board[i][j][k] == player for i in range(4) for j in range(4) for k in range(4) if i == j == k):
        return True

    if all(board[i][j][k] == player for i in range(4) for j in range(4) for k in range(4) if i == j == 3 - k):
        return True

    if all(board[i][j][k] == player for i in range(4) for j in range(4) for k in range(4) if i == 3 - j == k):
        return True

    if all(board[i][j][k] == player for i in range(4) for j in range(4) for k in range(4) if 3 - i == j == k):
        return True

    # No wins found
    return False

def check_draw(board):
    return all(cell != ' ' for i in range(4) for j in range(4) for k in range(4) for cell in board[i][j][k])

def player_move(board, player):
    while True:
        try:
            move = int(input(f"Player {player}, enter your move (1-64): ")) - 1
            i, j, k = divmod(move, 16), divmod(move % 16, 4), move % 4
            if 0 <= i < 4 and 0 <= j < 4 and 0 <= k < 4 and board[i][j][k] == ' ':
                board[i][j][k] = player
                break
            else:
                print("Invalid move. Try again.")
        except ValueError:
            print("Invalid input. Enter a number (1-64).")

def computer_move(board, computer, player):
    # Implement an intelligent computer move
    # (e.g., block player's winning moves and aim for a winning move)
    for i in range(4):
        for j in range(4):
            for k in range(4):
                if board[i][j][k] == ' ':
                    board[i][j][k] = computer
                    if check_win(board, computer):
                        return
                    board[i][j][k] = ' '  # Undo the move

    # If there's no immediate win or block, make a random move
    available_moves = [i * 16 + j * 4 + k for i in range(4) for j in range(4) for k in range(4) if board[i][j][k] == ' ']
    if available_moves:
        move = random.choice(available_moves)
        i, j, k = divmod(move, 16), divmod(move % 16, 4), move % 4
        board[i][j][k] = computer

def main():
    board = [[[[' ' for _ in range(4)] for _ in range(4)] for _ in range(4)] for _ in range(4)]
    
    # Player chooses whether to go first or second
    player_choice = input("Do you want to go first? (y/n): ").lower()
    
    if player_choice == 'y':
        player = 'X'
        computer = 'O'
    else:
        player = 'O'
        computer = 'X'

    while True:
        print_board(board)
        
        if player == 'X':
            player_move(board, player)
        else:
            computer_move(board, computer, player)

        if check_win(board, player):
            print_board(board)
            print(f"Player {player} wins!")
            break

        if check_draw(board):
            print_board(board)
            print("It's a draw!")
            break

        player = 'X' if player == 'O' else 'O'

if __name__ == "__main__":
    main()
```

```python
# Q-12

import numpy as np

def median(arr):
    arr = arr.flatten()
    arr = np.sort(arr)
    if len(arr) % 2 == 0:
        return (arr[int(len(arr)/2)] + arr[int(len(arr)/2) - 1]) / 2
    else:
        return arr[int(len(arr)/2)]

def mode(arr):
    
    #return_counts will return the no of counts of a num in array
    
    values, counts = np.unique(arr, return_counts=True)
    index = np.argmax(counts)
    return values[index]


arr1 = np.array([11, 22, 35, 46, 5])

arr2 = np.array([[1, 24, 3], [47, 53, 6]])

arr3 = np.array([1, 22, 3, 43, 4, 44, 4, 5, 53, 53, 45])


print("\nArray 1 Values \n")
print(arr1)
print("\nMedian of Array 1 \n")
print(median(arr1))
print("\nMode of Array 1 \n")
print(mode(arr1))


print("\nArray 2 Values \n")
print(arr2)
print("\nMedian of Array 2 \n")
print(median(arr2))
print("\nMode of Array 2 \n")
print(mode(arr2))

print("\nArray 3 Values \n")
print(arr3)
print("\nMedian of Array 3 \n")
print(median(arr3))
print("\nMode of Array 3 \n")
print(mode(arr3))

```

```python
# Q-13

def enhanced_median(arr, axis=None):
    
    arr = np.array(arr)
    
    if axis is None:        
        return np.median(arr)
    elif axis == 0:
        
        return np.median(arr, axis=0)
    elif axis == 1:
        
        return np.median(arr, axis=1)
    else:
        raise ValueError("Axis must be 0 in column  or 1 in row")

def enhanced_mode(arr, axis=None):
    
    arr = np.array(arr)
    
    if axis is None:
        
        return np.apply_along_axis(lambda x: np.argmax(np.bincount(x)), axis=0, arr=arr)
    elif axis == 0:
        
        return np.apply_along_axis(lambda x: np.argmax(np.bincount(x)), axis=0, arr=arr)
    elif axis == 1:
        
        return np.apply_along_axis(lambda x: np.argmax(np.bincount(x)), axis=1, arr=arr)
    else:
        raise ValueError("Axis must be 0  in column or 1 in row)")


arr = np.array([[61, 24, 33, 44],
                [53, 2, 32, 45],
                [15, 25, 33, 42]])


arr2 = np.array([[13, 42, 33, 42],
                [32, 22, 13, 24],
                [15, 25, 43, 34]])

arr2 = np.array([[17, 32, 33, 42],
                [32, 24, 11, 24],
                [75, 27, 44, 54]])


print("Array 1\n")

column_medians = enhanced_median(arr, axis=0)
print("Median in Column", column_medians)


row_medians = enhanced_median(arr, axis=1)
print("Median in Row", row_medians)


column_modes = enhanced_mode(arr, axis=0)
print("Mode in Column", column_modes)


row_modes = enhanced_mode(arr, axis=1)
print("Mode in Row", row_modes)



print("\nArray 2\n")

column_medians = enhanced_median(arr2, axis=0)
print("Median in Column", column_medians)


row_medians = enhanced_median(arr2, axis=1)
print("Median in Row", row_medians)


column_modes = enhanced_mode(arr2, axis=0)
print("Mode in Column", column_modes)


row_modes = enhanced_mode(arr2, axis=1)
print("Mode in Row", row_modes)



print("\nArray 3\n")
column_medians = enhanced_median(arr3, axis=0)
print("Median in Column", column_medians)


row_medians = enhanced_median(arr3, axis=1)
print("Median in Row", row_medians)


column_modes = enhanced_mode(arr3, axis=0)
print("Mode in Column", column_modes)


row_modes = enhanced_mode(arr3, axis=1)
print("Mode in Row", row_modes)
```
